```python
import cv2 as cv
import numpy as np
import matplotlib.pyplot as plt
import tensorflow as tf

img_path = r"C:\Users\maksym.kashchuk\Desktop\auto-os-ml\datasets\CuLane\train-validation\driver_161_90frame\06031152_0826.MP4\00000.jpg"
# img_path = r"C:\Users\maksym.kashchuk\Desktop\auto-os-ml\core\lane-worker\mask.jpg"

img = cv.imread(img_path)

# Перевіряємо, чи зображення було завантажено
assert img is not None, "Image file could not be read. Check the file path."

# Додаємо паддінг до 1024x1024, зберігаючи співвідношення сторін
desired_size = 1024
height, width, _ = img.shape

# Вираховуємо нові розміри зображення, зберігаючи співвідношення сторін
if height > width:
    new_height = desired_size
    new_width = int((width / height) * desired_size)
else:
    new_width = desired_size
    new_height = int((height / width) * desired_size)

# Масштабуємо зображення
resized_img = tf.image.resize(img, (new_height, new_width))
# resized_img = cv.resize(img, (new_width, new_height))

# Обчислюємо паддінг для досягнення бажаного розміру
top_padding = (desired_size - new_height) // 2
bottom_padding = desired_size - new_height - top_padding
left_padding = (desired_size - new_width) // 2
right_padding = desired_size - new_width - left_padding

img_tensor = tf.convert_to_tensor(img, dtype=tf.float32)

# Додаємо паддінг
desired_size = 1024

padded_img = tf.image.resize_with_pad(img_tensor, desired_size, desired_size,
                                      method=tf.image.ResizeMethod.BILINEAR)
# padded_img = cv.copyMakeBorder(resized_img, top_padding, bottom_padding, left_padding, right_padding, cv.BORDER_CONSTANT, value=(0, 0, 0))

cv.imshow('Image with Padding', padded_img.numpy().astype(np.uint8))

# # Чекаємо на натискання клавіші, щоб закрити вікно
cv.waitKey(0)
cv.destroyAllWindows()
```
